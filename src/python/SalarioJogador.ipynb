```python
#Realizando a conversao da data de anivesario do jogador para a idade
from datetime import date
today = date.today()
def getAge(birthday):
    return today.year - birthday.year - ((today.month,
                                          today.day) < (birthday.month,
                                                        birthday.day))
# # Aplicando funcao
dflimp.BIRTHDATE = dflimp.BIRTHDATE.map(lambda x: getAge(x))
```

```python
#Renomeando a coluna dia de aniversario para idade
dflimp.rename(columns={"BIRTHDATE":"age"},inplace = True)
```
