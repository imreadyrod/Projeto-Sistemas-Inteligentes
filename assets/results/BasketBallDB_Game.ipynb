```python
#Determinando os times participantes da temporada  

df_equipes = df_temp2020["TEAM_ID_HOME"].groupby(df_temp2020["TEAM_NAME_HOME"]).count()
print(df_equipes)
```

```python
#Determinando a quantidade de vitorias e derrotas por cada time


df_ganhos = df_temp2020[df_temp2020["WL_HOME"] == 'W'].groupby(df["TEAM_NAME_HOME"]).count().sort_values(by="WL_HOME",ascending = False)
df_ganhos


```

```python
df_perdas = df_temp2020[df_temp2020["WL_HOME"] == 'L'].groupby(df["TEAM_NAME_HOME"]).count().sort_values(by="WL_HOME",ascending = False)
df_perdas
```

```python

df_empates = df_temp2020[df_temp2020["WL_HOME"] ==" "].groupby(df["TEAM_NAME_HOME"]).count().sort_values(by="WL_HOME",ascending = False)
df_empates
```

```python
#realizando gráfico de barras para a quantidade de vitorias de cada time
#Para isso devemos saber qual a quantidade de vitorias cada time teve
#Entao transformamos a WL_HOME em string
#A partir disso podemos filtrar apenas as vitorias
df_temp2020['WL_HOME'] = df_temp2020['WL_HOME'].astype('str')
df_temp2020.dtypes
#df_graficobarra = df_temp2020.loc[(df_temp2020['TEAM_NAME_HOME'])&(df_temp2020['WL_HOME']=='W')&(df_temp2020['SEASON']==2020)]
#df_graficobarra
```

```python
#Filtrando apenas as linhas que contenham vitorias
win_temp2020 = df_temp2020[df_temp2020['WL_HOME'].str.contains('W')]
win_temp2020
```

```python
#Criando um novo dataframe com apenas o TEAM_NOME e os WL_HOME
df_teamresult = win_temp2020.loc[:,['TEAM_NAME_HOME','WL_HOME']]
df_teamresult
```

```python
#Realizando a contagem com groupby
df_teamresultfinal = df_teamresult.groupby('TEAM_NAME_HOME').count().sort_values(by='WL_HOME')
df_teamresultfinal
```

```python
#Realizando o gráfico de barras
df_teamresultfinal.plot(kind='bar')
```
